# Handling null value

```python
# checking null data
counts = df.isna().sum().sort_values(ascending=False)
# check the percentage of empty data
percentages = round(df.isna().mean() * 100, 2).sort_values(ascending=False)
null_values = pd.concat([counts, percentages], axis=1, keys=["count", "%"])
null_values
```

```python
df['company'] = df['company'].fillna(0)
df['agent'] = df['agent'].fillna(0)
df['children'] = df['children'].fillna(0)
df['city'] = df['city'].fillna('unknown')
```

```python
# checking null data
counts = df.isna().sum().sort_values(ascending=False)
# check the percentage of empty data
percentages = round(df.isna().mean() * 100, 2).sort_values(ascending=False)
null_values = pd.concat([counts, percentages], axis=1, keys=["count", "%"])
null_values
```

# Replace inappropriate values

```python
df['meal'].value_counts()
```

```python
# replace the 'Undefined' value in meal columns to 'No Meal'
df = df.replace({'meal': {'Undefined':'No Meal'}})
df['meal'].value_counts()
```

```python
# replace few columns data type from float64 to int64
df[['children', 'agent', 'company', 'days_in_waiting_list', 'required_car_parking_spaces', 'total_of_special_requests']] = df[['children', 'agent', 'company', 'days_in_waiting_list', 'required_car_parking_spaces', 'total_of_special_requests']].astype('int64')
```

```python
df.info()
```

```python
df_clean = df.copy()
```

# Drop unnecessary data

```python
# checking the total guest and total stay duration, if 0 
df_clean['total_guest'] = df_clean['adults'] + df_clean['children'] + df_clean['babies']
df_clean['total_stay_duration'] = df_clean['stays_in_weekend_nights'] + df_clean['stays_in_weekdays_nights']

print(f'Number of data rows with 0 total guest: {df_clean[df_clean["total_guest"] == 0].shape[0]} out of {len(df_clean)} data')
print(f'Number of data rows with 0 total stay duration: {df_clean[df_clean["total_stay_duration"] == 0].shape[0]} out of {len(df_clean)} data')
```

```python
df_clean_final = df_clean[(df_clean['total_guest'] > 0) & (df_clean['total_stay_duration'] > 0)]

print(f'Number of data before filtering : {len(df_clean)} rows')
print(f'Number of data after filtering : {len(df_clean_final)} rows')
```

```python
df_fix = df_clean_final.copy()
```

# EDA

```python
df_fix.groupby(['hotel', 'arrival_date_year', 'arrival_date_month']).agg({'is_canceled' : 'count'})
```

```python
# create aggregate table to show total hotel booking per month based on hotel category
df_table = df_fix.groupby(['hotel', 'arrival_date_year', 'arrival_date_month']).agg({'is_canceled' : 'count'}).reset_index()
```

```python
# create amount of year based on hotel category and month
df_year = df_table.groupby(['hotel', 'arrival_date_month']).agg({'arrival_date_year':'count'})
df_table = df_table.merge(df_year, on=['hotel', 'arrival_date_month'])
df_table
```

```python
# create sum of booking based on hotel category and month
df_booking_sum = df_table.groupby(['hotel', 'arrival_date_month']).agg({'is_canceled' : 'sum'})
df_table = df_table.merge(df_booking_sum, on=['hotel', 'arrival_date_month'])
df_table = df_table.rename(columns = {'arrival_date_year_x' : 'arrival_date_year', 'is_canceled_x' : 'num_of_booking', 'arrival_date_year_y' : 'amount_of_year', 'is_canceled_y' : 'sum_of_booking'})


df_table['avg_num_booking'] = round(df_table['sum_of_booking']/df_table['amount_of_year'])
df_table
```

```python
#create percentage column
df_month_percent = df_table[['hotel', 'arrival_date_month', 'avg_num_booking']]

#drop duplicate data
df_month_percent = df_month_percent.drop_duplicates()

#summarize total of avg num booking based on hotel type
df_part = df_month_percent.groupby('hotel').agg({'avg_num_booking':'sum'})
df_month_percent = df_month_percent.merge(df_part, on='hotel')
df_month_percent = df_month_percent.rename(columns = {'avg_num_booking_x' : 'avg_num_booking', 'avg_num_booking_y' : 'total_avg_booking'})

#create percentage column based on avg num booking per hotel type
df_month_percent['percentage'] = round((df_month_percent['avg_num_booking']/df_month_percent['total_avg_booking'])*100, 2)
df_month_percent =  df_month_percent.replace({'arrival_date_month': {'January':1, 'February':2, 'March':3, 'April':4, 'May':5, 'June':6, 
                                           'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}})
df_month_percent.sort_values('arrival_date_month', ascending=True)
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
plt.title("Average Number of Hotel Bookings per Month Over Year\nBased on Hotel Types", fontsize=15, color='black', weight='bold', pad=80)
plt.text(x=-1.5, y=5250, s="June, July, and August were the months with the most hotel bookings. City hotels reached the highest average number of hotel bookings of 11.18% in July and resort hotels reached 9.82% in June.\nThis can happen because that month is the school vacation season. An increase in the number of bookings also occurred in December with city hotels by 10.32% and resort hotels by 9.61%.\nThis can occur due to the Christmas and New Year holidays.", fontsize=12, fontstyle='italic')
sns.barplot(x='arrival_date_month', y='avg_num_booking', data=df_month_percent, hue='hotel', edgecolor='black', palette='muted')

plt.xlabel('Month(s)', fontsize=11)
plt.xticks(np.arange(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.ylabel('Average Number of Booking', fontsize=11)
plt.ylim(0, 5000)

plt.bar_label(ax.containers[0], padding=2)
plt.bar_label(ax.containers[1], padding=2)

plt.axvline(4.5, ls='--', color='green')
plt.axvline(7.5, ls='--', color='green')
plt.stackplot(np.arange(4.5,8.5), [[5000]], color='green', alpha=0.3)
plt.text(x=5.3, y=4750, s='School Vacation Season', fontsize=14, color='green', va='center')

plt.axvline(10.5, ls='--', color='red')
plt.axvline(11.5, ls='--', color='red')
plt.stackplot(np.arange(10.5,12.5), [[5000]], color='pink', alpha=0.3)
plt.text(x=10.65, y=4650, s='Christmas\nNew Year\n Holiday', fontsize=14, color='red', va='center')

plt.bar_label(ax.containers[0], ['5.88%', '5.59%', '5.07%', '6.71%', '8.76%', '10.18%', '11.18%', '10.75%', '7.33%', '8.15%', '10.07%', '10.32%'], label_type='center', color='white', weight='bold', fontsize=8)
plt.bar_label(ax.containers[1], ['6.58%', '7.09%', '5.91%', '8.43%', '9.04%', '9.82%', '9.63%', '8.25%', '8.27%', '8.87%', '8.48%', '9.61%'], label_type='center', color='white', weight='bold', fontsize=8)

sns.despine()
plt.tight_layout()
plt.savefig('avg_num_booking.png', dpi=200)
```
